# Ranking Plot

## Ordered Bar Chart

### Step 1 : Initialize Global Setting

```python
#Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.patches as patches

# Setting up the params
large = 22
med =16
small= 12

params = {
    'axes.titlesize':large,
    'legend.fontsize':med,
    'figure.figsize':(16,10),
    'axes.labelsize':med,
    'xtick.labelsize':med,
    'ytick.labelsize':med,
    'figure.titlesize':large
    
}

#override default parameters of plot

plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
```

### Step 2 :  Import dataset 

```python
df=pd.read_csv("https://raw.githubusercontent.com/mukesh-ranjan/LearnDataAnalysisFromScratch/master/Data/mpg.csv")
df
```

### Step 3 : Prepare Data

```python
processeddf=df[['cty','manufacturer']].groupby('manufacturer').apply(lambda x:x.mean())
processeddf.sort_values('cty',inplace=True)
processeddf.reset_index(inplace=True)
```

### Step 4 : Draw Plot

```python
fig,ax=plt.subplots(figsize=(16,10),facecolor='white',dpi=80)

ax.vlines(x=processeddf.index,ymin=0,ymax=processeddf.cty,color='firebrick',alpha=0.7,linewidth=20)

for i,cty in enumerate(processeddf.cty):
    ax.text(i,cty+0.5,round(cty,1), horizontalalignment='center')
    
ax.set_title('Bar Chart For City Mileage',fontdict={'size':22})
ax.set(ylabel='Miles Per Gallon',ylim=(0,30))

plt.xticks(processeddf.index,processeddf.manufacturer.str.upper(),rotation=60, horizontalalignment='right',fontsize=small)

p1 = patches.Rectangle((.57, -0.005), width=.33, height=.13, alpha=.1, facecolor='green', transform=fig.transFigure)
p2 = patches.Rectangle((.124, -0.005), width=.446, height=.13, alpha=.1, facecolor='red', transform=fig.transFigure)

fig.add_artist(p1)
fig.add_artist(p2)
plt.show()
```
