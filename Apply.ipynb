# Apply Method

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp=np.arange(90,150)
temp
```

```python
df=pd.DataFrame(temp,columns=['Fahrenheit'])
df
```

```python
def convertToCelsius(ftemp):
    return((ftemp-32)*5/9)
```

```python
df['Celsius']=df['Fahrenheit'].apply(convertToCelsius)
df
```

```python
plt.plot(df['Fahrenheit'],df['Celsius'])
```
