# Let's Understand Pandas Filter

```python
import pandas as pd
```

```python
df=pd.read_csv("Data/grocery.csv")
df
```

```python
df_filter=df.filter(items=["Type","Items"])
df_filter
```

```python
df_rgex=df.filter(regex='e$',axis=1)
df_rgex
```

```python
df.set_index('Type',inplace=True)
```
