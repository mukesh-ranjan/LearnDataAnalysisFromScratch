# Handling Missing Data

```python
import pandas as pd
```

```python
df=pd.read_csv('Data/wine.csv')
```

```python
df
```

#### 1. Check for Missing Data

```python
df['Alcohol'].isna()
```

```python
df.isna()
```

#### 2. Get Count Of Missing Data 

```python
df.isnull().sum()
```

#### 3. Filling Missing Data

##### Forward Fill

```python
df.fillna(method='ffill',inplace=True)
```

##### Backward Fill

```python
df.fillna(method='bfill')
```

#### 4. Interpolate Missing Value

```python
df
```

```python
df.interpolate(method='pchip')
```

#### 5. When to Drop missing value

```python
df.dropna()
```
