```python
import numpy as np
import pandas as pd

```

# Join

```python
df1 = pd.DataFrame({
                    'Key': ['X', 'Y'],
                   'A': ['A1', 'A2'],

                   'B': ['B1', 'B2']})
df1.set_index("Key",inplace=True)
df1
```

```python
df2 = pd.DataFrame({
                    'Key': ['X', 'Z'],
                   'A': ['A3', 'A4'],

                   'C': ['C1', 'C2']})
df2.set_index("Key",inplace=True)
df2
```

```python
df1.join(df2, lsuffix='_df1', rsuffix='_df2',how='left')
```

```python
df1.join(df2, lsuffix='_df1', rsuffix='_df2',how='right')
```

```python
df1.join(df2, lsuffix='_df1', rsuffix='_df2',how='inner')
```

```python
df1.join(df2, lsuffix='_df1', rsuffix='_df2',how='outer')
```

# Merge

```python
df1.merge(df2,how='left',left_on='A',left_index=True,right_index=True)
```

# Concate

```python
pd.concat([df1,df2],axis=0, join='outer')
```
