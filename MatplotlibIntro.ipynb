# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Learn By Doing Exercises

#### 1. How to draw Curve using matplotlib

```python
X=np.linspace(0,2*np.pi,100)
```

```python
y=np.sin(X)
```

```python
plt.plot(X,y)
```

#### 2. How to draw multiple curve on the same plot

```python
z=np.cos(X)
```

```python
plt.plot(X,y)
plt.plot(X,z)
```
