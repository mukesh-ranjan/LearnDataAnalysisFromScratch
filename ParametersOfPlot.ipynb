```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/mukesh-ranjan/LearnDataAnalysisFromScratch/master/Data/mtcars.csv")
```

```python
df
```

```python
plt.figure(figsize=(12,10),dpi=80)
```

```python
sns.heatmap(df.corr(),xticklabels=df.corr().columns,yticklabels=df.corr().columns,cmap='BuPu',center=0,annot=True)
plt.title('Heatmap for mtcars')
```

```python
large = 25
med=16 
small=12
```

```python
params = {'axes.titlesize':large,
          'legend.fontsize':med,
          'figure.figsize':(16,10),
          'axes.labelsize':med,
           'xtick.labelsize':med,
           'ytick.labelsize':med,
         'figure.titlesize':large}
plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style("white")
```
