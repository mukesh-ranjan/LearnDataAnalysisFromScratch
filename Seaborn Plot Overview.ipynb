```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rng=np.random.RandomState(0)
x=np.linspace(0,10,500)
y=np.cumsum(rng.randn(500,6),0)
```

```python
plt.plot(x,y)
plt.legend('ABCDEF',ncol=2 ,loc='upper left')
```

```python
sns.set()
```
