Q1 : What percentage of the order total does each product represent ?

```python
ototal = df.groupby('order')["ext price"].sum().rename("Order_Total").reset_index()
ototal
```

```python
df_merge=df.merge(ototal)
df_merge
```

```python
df_merge["%OfOrderByProduct"]=df_merge["ext price"]/ df_merge["Order_Total"]
df_merge
```

```python
df
```

```python
df["Order_Total"]=df.groupby('order')["ext price"].transform('std')
df
```

```python
df["%ByProductOrder"]=df["ext price"]/df["Order_Total"]
df
```
