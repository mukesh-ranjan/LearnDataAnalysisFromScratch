# Groupby

```python
import pandas as pd
```

```python
df=pd.read_csv("Data/wine.csv")
df
```

#### 1. How To Create Group Object

```python
df_wine=df.groupby(["Wine"])
df_wine
```

#### 2. Frequently Used Groupby Operations

```python
df_wine.first()
```

```python
df_wine.last()
```

```python
df_wine['Alcohol'].max()
```

```python
df_wine['Alcohol'].min()
```

```python
df_wine['Alcohol'].mean()
```

#### 3. Iterating Groups

```python
df_wine.groups
```

```python
for wine,wine_df in df_wine:
    print(wine)
    print(wine_df)
```

#### 4. How to Get Information about a Specific Group

```python
df_wine.get_group(1)
```

#### 5. Getting Detailed View of the Groups 

```python
df_wine.describe()
```

#### 6. Check Various functions associated with groupby object

```python
dir(df_wine)
```

#### 7. Check The Length Of Group

```python
len(df_wine.get_group(1))
```

```python
len(df_wine)
```
